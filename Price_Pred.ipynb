```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as mp
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from xgboost import XGBRegressor
import pickle
```

```python
dataset = pd.read_csv("/content/BostonHousing.csv")
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
x = dataset.drop('price',axis=1)
y = dataset['price']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=32)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
model_prediction = model.predict(x_train)
model_prediction
```

```python
s1 = metrics.r2_score(model_prediction, y_train)
print("R2 Score:",s1)
s2 = metrics.mean_absolute_error(model_prediction, y_train)
print("Mean Absolute Error:",s2)
```

```python
model2 = XGBRegressor()
model2.fit(x_train, y_train)
```

```python
model_prediction = model2.predict(x_train)
model_prediction
```
